```python
# STEP 1: IMPORTS

import pandas as pd
from sklearn.ensemble import IsolationForest


# STEP 2: LOAD DATASET

df = pd.read_csv("/content/drive/MyDrive/Major Project/Model training/train_scaled_processed.csv")
```

```python
# STEP 3: PREPROCESSING

X_iso = df.drop(columns=['Fatigue'], errors='ignore')
X_iso = X_iso.dropna(axis=1, how='all')
X_iso = X_iso.fillna(X_iso.mean())
```

```python
# STEP 4: ISOLATION FOREST

iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.1,
    random_state=42
)

iso_labels = iso_forest.fit_predict(X_iso)
```

```python
# STEP 5: ADD RESULTS

df['IsolationForest_Anomaly'] = iso_labels

df['IsolationForest_Label'] = df['IsolationForest_Anomaly'].apply(
    lambda x: "Anomalous User" if x == -1 else "Normal User"
)
```

```python
# STEP 6: RESULTS

print("\nIsolation Forest Distribution:")
print(df['IsolationForest_Anomaly'].value_counts())

num_anomalies = (df['IsolationForest_Anomaly'] == -1).sum()
total = len(df)

print("Number of Anomalies:", num_anomalies)
print("Anomaly %:", (num_anomalies / total) * 100)

# STEP 7: SAVE OUTPUT

df.to_csv("isolation_forest_output.csv", index=False)
```
